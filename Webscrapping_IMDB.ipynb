**Load first page**

```python
driver=gs.Chrome()
driver.get("https://www.imdb.com/title/tt4154664/reviews")
load_more=driver.find_element(By.XPATH,
                              "/html/body/div[2]/div/div[2]/div[3]/div[1]/section/div[2]/div[4]/div/button")
load_more.click()
```

**Extracting reviews and ratings from the HTML.**

```python
reviews_complete=driver.find_element(By.XPATH,
"/html/body/div[2]/div/div[2]/div[3]/div[1]/section/div[2]/div[2]")
print(reviews_complete.text)
```

**Separate lists for data scrapped.**

```python
reviews=[]
ratings=[]
for i in range(1,10):
  path=f"/html/body/div[2]/div/div[2]/div[3]/div[1]/section/div[2]/div[2]/div[{i}]"
  review=driver.find_element(By.XPATH,path).text.split("\n")
  ratings.append(review[0])
  reviews.append(review[3])

review
```

**Saving the extracted data to a data frame.**


```python
import pandas as pd
```

```python
df=pd.DataFrame({"reviews":reviews,"ratings":ratings})
```

```python
df.to_csv("reviews.csv")
```
